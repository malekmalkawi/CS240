```python
resp = nsfg.ReadFemResp()
```

```python
splitting = resp['hieduc'] < 10
splitting2 = resp['hieduc'] > 10
school = resp[splitting]
university = resp[splitting2]
```

```python
width = 0.5

school_pmf = thinkstats2.Pmf(school['hieduc'], label='school')
univeristy_pmf = thinkstats2.Pmf(university['hieduc'],label = 'university')

thinkplot.PrePlot(2)
thinkplot.Hist(school_pmf, align='right', width=width)
thinkplot.Hist(univeristy_pmf, align='left', width=width)
thinkplot.Config(xlabel='education', ylabel='PMF')
```

```python
thinkplot.preplot(2)
first = thinkstats2.Cdf(school.agemomb1, label='school')
other = thinkstats2.Cdf(university.agemomb1,label = 'university')
thinkplot.Cdfs([first,other])
thinkplot.Config(xlabel='both variable', ylabel='CDF')
```

```python
first_pmf = thinkstats2.Pmf(school['agemomb1'], label='first')
other_pmf = thinkstats2.Pmf(university['agemomb1'],label = 'other')

a=first_pmf - other_pmf
thinkplot.PrePlot(1)
thinkplot.Hist(a, align='right', width=width)
#thinkplot.Hist(other_pmf, align='left', width=width)
thinkplot.Config(xlabel='difference for ages of mothers at first birth', ylabel='PMF')
```

```python
school['agemomb1']
university['agemomb1']

thinkplot.preplot(2)
first = thinkstats2.Cdf(school.agemomb1, label='school')
other = thinkstats2.Cdf(university.agemomb1,label = 'university')
thinkplot.Cdfs([first,other])
thinkplot.Config(xlabel=' ages of mothers at first birth', ylabel='CDF')
```

```python
thinkplot.PrePlot(2) 
xs, ys = thinkstats2.NormalProbability(school['agemomb1']<18)
thinkplot.Plot(xs, ys, label='all live')

xs, ys = thinkstats2.NormalProbability(university['agemomb1']<18)
thinkplot.Plot(xs, ys, label='full term')
thinkplot.Config(title='Normal probability plot',
                 xlabel='Standard deviations from mean',
                 ylabel='Birth weight (lbs)')
```

```python
pmf = thinkstats2.Pmf(school['agemomb1'])
pmf2 = thinkstats2.Pmf(university['agemomb1'])

smaller1 = []
larger1 =[]
smaller2 = []
larger2 =[]

for i in pmf:
    if i < 18:
        smaller1.append(pmf[i])
    else :
        larger1.append(pmf[i])
for i in pmf2:
    if i < 18:
        smaller2.append(pmf2[i])
    else:
        larger2.append(pmf2[i])
        
print(smaller1)
print(smaller2)
print(larger1)
print(larger2)
```
